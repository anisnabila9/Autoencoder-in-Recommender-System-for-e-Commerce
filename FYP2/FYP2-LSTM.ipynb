```python
def precision_at_k(predicted, actual, k):
    correct_predictions = 0
    for user_id in range(len(predicted)):
        top_k_items = np.argsort(predicted[user_id].flatten())[-k:][::-1]
        actual_items = np.argsort(actual[user_id].flatten())[-k:][::-1]
        correct_predictions += len(set(top_k_items) & set(actual_items))
    return correct_predictions / (len(predicted) * k)

def recall_at_k(predicted, actual, k):
    correct_predictions = 0
    for user_id in range(len(predicted)):
        top_k_items = np.argsort(predicted[user_id].flatten())[-k:][::-1]
        actual_items = np.argsort(actual[user_id].flatten())[-k:][::-1]
        correct_predictions += len(set(top_k_items) & set(actual_items))
    return correct_predictions / (len(predicted) * k)

k = 10
precision = precision_at_k(predicted_preferences, test_interaction_matrix, k)
recall = recall_at_k(predicted_preferences, test_interaction_matrix, k)

print(f"Precision@{k}: {precision}")
print(f"Recall@{k}: {recall}")
```
